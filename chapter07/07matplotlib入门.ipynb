```python
import random
from cProfile import label

import matplotlib.pyplot as plt
from matplotlib import figure
from matplotlib.axes import Axes
from openai.resources.fine_tuning.alpha import alpha

# 绘制折线图
x = range(1, 11)
y = [6, 2, 9, 8, 10, 1, 5, 4, 7, 6]

# 折线图
plt.plot(x, y)
plt.show()
```

```python
# 中文字体
plt.rcParams['font.sans-serif'] = ['PingFang SC', 'Heiti SC', 'STHeiti', 'Arial Unicode MS']
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示为方块

x = [i for i in range(1, 25)]
y = [random.randint(10, 15) for i in x]
z = [random.randint(10, 15) for i in x]


plt.figure(figsize=(10, 5)) # 设置画布大小，单位为英寸
plt.plot(x, y, label='北京') # 如果没有画布，会自动创建一个画布
plt.plot(x, z, label='上海')

# 设置折现图的详细信息
plt.title('气温变化折线图', fontsize=13)

plt.xlabel('时间') # x轴标签
plt.ylabel('温度') # y轴标签

plt.xticks(x[::2], fontsize=13) # x轴刻度
plt.yticks(y[::2], fontsize=13) # y轴刻度

plt.grid(linestyle='-.', linewidth=0.5, color='grey', alpha=0.5)

plt.legend(loc='upper right', fontsize=13)

plt.show()
```

```python
# 创建子图
figure, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 6), dpi=100)

# 中文字体
plt.rcParams['font.sans-serif'] = ['PingFang SC', 'Heiti SC', 'STHeiti', 'Arial Unicode MS']
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示为方块

countries = ['中国', '美国', '印度', '加拿大', '伊拉克', '沙特', '伊朗', '委内瑞拉']
values = [35, 23, 18, 21, 56, 78, 66, 70]

axes1: Axes = axes[0]
axes1.set_title('世界石油储备', fontsize=13) # 设置标题
axes1.bar(countries, values, 0.6, color='r') # 绘制柱状图
axes1.set_xlabel('国家') # 设置x轴的标签
axes1.set_ylabel('石油储备') # 设置y轴标签
axes1.grid(linestyle='-.', linewidth=0.5, color='grey', alpha=0.7) # 网格线

countries = ['印度', '中国', '美国', '印尼', '巴基斯坦', '尼日利亚', '巴西', '俄罗斯', '其他']
values = [14.51, 14.09, 3.4, 2.83, 2.51, 2.33, 2.12, 1.44, 20]

axes2: Axes = axes[1]
axes2.pie(values, labels=countries, autopct='%1.1f%%')
axes2.set_title('世界人口比例', fontsize=13)
axes2.legend(loc='lower center', fontsize=13, ncol=3, bbox_to_anchor=(0.5, -0.15), fancybox=True) # 设置图例，ncol：每行显示的个数， bbox_to_anchor: 图例的位置

plt.savefig('resources/result.png')

```
