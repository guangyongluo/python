```python
import pandas as pd

# 读取数据
df = pd.read_csv('resources/sales.csv', usecols=['订单号', '产品类别', '产品名称', '销售数量', '单价'])
print(df)

# 数据处理
df['商品金额'] = df['销售数量'] * df['单价']
print(df)

# 写入数据
df.to_csv('resources/sales_01.csv', index=False)
```
