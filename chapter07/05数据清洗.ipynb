```python
import pandas as pd

pd.set_option('display.max_rows', 30)

df = pd.read_csv('resources/sales.csv')

df.isnull() # 查看缺失值

df.dropna() # 删除缺失值所在行
df.dropna(axis=1) # 删除缺失值所在列
df.fillna('--') # 填充缺失值
df.ffill() # 使用上一行填充缺失值
df.bfill() # 使用下一行填充缺失值


df.duplicated() # 查看重复值(所有的列都一样才会判断是重复的)
df.duplicated(subset=['订单号']) # 查看指定列是否为重复值
df.drop_duplicates(subset=['订单号'],keep='first') # 保留第一个，删除其他的重复值

df[df['单价'] < 0] # 查看异常数据
df.drop(df[['单价'] < 0].index)

df['单价'] = df['单价'].abs()


# 数据格式处理
df['订单日期'] = df['订单日期'].str.replace('/','-')


df.sort_values('单价', ascending=False)
df.sort_values('单价')

df.sort_values(['单价', '销售数量'], ascending=[True, False])
```
